## 1. Kütüphane ve Veri Yükleme

Pandas ve Numpy kütüphanelerini import ettik ve weather_data.csv dosyasını yükledik.

```python
import pandas as pd
import numpy as np

# CSV dosyasını okuyoruz
df = pd.read_csv('weather_data.csv')

print("Veri seti başarıyla yüklendi!")
print(f"Veri seti boyutu: {df.shape}")
print(f"Sütun isimleri: {df.columns.tolist()}")
```

## 2. Veriyi Keşfetme

Veri setinin ilk 5 satırı, son 5 satırı ve istatistiksel özeti görüntülenmiştir.

```python
# İlk 5 satırı gösteriyoruz
print("İlk 5 satır:")
print(df.head())
print("\n" + "="*70 + "\n")
```

```python
# Son 5 satırı gösteriyoruz
print("Son 5 satır:")
print(df.tail())
print("\n" + "="*70 + "\n")
```

```python
# İstatistiksel özet
print("Veri seti istatistiksel özeti:")
print(df.describe())
```

## 3. Sütun Seçimi

Sadece Date, City ve Temperature sütunlarını seçerek yeni bir görünüm oluşturduk.

```python
# Sadece Date, City ve Temperature sütunlarını seçiyoruz
selected_columns = df[['Date', 'City', 'Temperature']]
print("Seçilen sütunlar (İlk 10 satır):")
print(selected_columns.head(10))
print("\n" + "="*70 + "\n")
```

```python
# City ve Temperature kolonlarını birlikte listeleyiyoruz
city_temp = df[['City', 'Temperature']]
print("Şehir ve Sıcaklık birlikte (İlk 15 satır):")
print(city_temp.head(15))
```

## 4. Basit Filtreleme

Belirli koşullara göre verileri filtreledik.

```python
# Sıcaklık 30 derecenin üzerinde olan kayıtlar
temp_above_30 = df[df['Temperature'] > 30]
print("Sıcaklık 30°C'nin üzerinde olan kayıtlar:")
print(f"Toplam kayıt sayısı: {len(temp_above_30)}")
print(temp_above_30.head(10))
print("\n" + "="*70 + "\n")
```

```python
# Bursa şehrine ait kayıtlar
bursa_records = df[df['City'] == 'Bursa']
print("Bursa şehrinin tüm kayıtları:")
print(f"Toplam kayıt sayısı: {len(bursa_records)}")
print(bursa_records.head(10))
```

## 5. Mantıksal Operatörler ile Filtreleme

AND ve OR operatörleri kullanarak karmaşık filtrelemeler yaptık.

```python
# İstanbul VE Nem > 60
istanbul_high_humidity = df[(df['City'] == 'İstanbul') & (df['Humidity'] > 60)]
print("İstanbul VE Nem > 60 olan kayıtlar:")
print(f"Toplam kayıt sayısı: {len(istanbul_high_humidity)}")
print(istanbul_high_humidity.head(10))
print("\n" + "="*70 + "\n")
```

```python
# Ankara VEYA Sıcaklık < 5
ankara_or_cold = df[(df['City'] == 'Ankara') | (df['Temperature'] < 5)]
print("Ankara VEYA Sıcaklık < 5°C olan kayıtlar:")
print(f"Toplam kayıt sayısı: {len(ankara_or_cold)}")
print(ankara_or_cold.head(10))
print("\n" + "="*70 + "\n")
```

```python
# Sıcaklık < 10 VEYA Nem > 70
cold_or_humid = df[(df['Temperature'] < 10) | (df['Humidity'] > 70)]
print("Sıcaklık < 10°C VEYA Nem > 70% olan kayıtlar:")
print(f"Toplam kayıt sayısı: {len(cold_or_humid)}")
print(cold_or_humid.head(10))
```

## 6. Sıralama (Sorting)

Verileri farklı sütunlara göre sıraladık.

```python
# Sıcaklığa göre büyükten küçüğe (azalan) sıralama - ilk 10 kayıt
sorted_by_temp = df.sort_values('Temperature', ascending=False)
print("Sıcaklığa göre büyükten küçüğe sıralanmış ilk 10 kayıt:")
print(sorted_by_temp[['Date', 'City', 'Temperature', 'Humidity']].head(10))
print("\n" + "="*70 + "\n")
```

```python
# Neme göre azalan şekilde sıralama
sorted_by_humidity = df.sort_values('Humidity', ascending=False)
print("Neme göre azalan şekilde sıralanmış ilk 10 kayıt:")
print(sorted_by_humidity[['Date', 'City', 'Humidity', 'Temperature']].head(10))
print("\n" + "="*70 + "\n")
```

```python
# Şehre göre artan şekilde sıralama
sorted_by_city = df.sort_values('City', ascending=True)
print("Şehre göre artan şekilde sıralanmış ilk 15 kayıt:")
print(sorted_by_city[['Date', 'City', 'Temperature', 'Humidity']].head(15))
```

## 7. Yeni Sütun Ekleme

Türetilmiş sütunlar (Temperature_F ve FeelsLike) ekledik.

```python
# Temperature_F sütunu ekliyoruz (Celsius'tan Fahrenheit'e dönüşüm)
# Formül: (Temperature * 9/5) + 32
df['Temperature_F'] = (df['Temperature'] * 9/5) + 32

print("Temperature_F sütunu eklendi (İlk 10 satır):")
print(df[['Date', 'City', 'Temperature', 'Temperature_F']].head(10))
print("\n" + "="*70 + "\n")
```

```python
# FeelsLike sütunu ekliyoruz
# Formül: FeelsLike = Temperature - (Humidity / 100)
df['FeelsLike'] = df['Temperature'] - (df['Humidity'] / 100)

print("FeelsLike sütunu eklendi (İlk 10 satır):")
print(df[['Date', 'City', 'Temperature', 'Humidity', 'FeelsLike']].head(10))
```

## 8. Gruplama (Grouping) ve Analiz

Şehirlere göre analiz yaptık.

```python
# Her şehirde kaç adet veri kaydı olduğunu hesaplıyoruz
city_counts = df['City'].value_counts().sort_index()
print("Her şehirdeki veri kaydı sayısı:")
print(city_counts)
print(f"\nToplam: {city_counts.sum()} kayıt")
print("\n" + "="*70 + "\n")
```

```python
# Şehirlere göre ortalama sıcaklık
avg_temp_by_city = df.groupby('City')['Temperature'].mean().sort_values(ascending=False)
print("Şehirlere göre ortalama sıcaklık (yüksekten düşüğe):")
print(avg_temp_by_city)
print(f"\nGenel ortalama sıcaklık: {df['Temperature'].mean():.2f}°C")
```

## 9. En Yüksek/Düşük Değer Analizi

Uç değerler (extremes) analiz ettik.

```python
# En yüksek sıcaklığın olduğu satır
max_temp_idx = df['Temperature'].idxmax()
max_temp_row = df.loc[max_temp_idx]

print("En yüksek sıcaklığın olduğu satır:")
print(max_temp_row)
print(f"\nEn yüksek sıcaklık: {max_temp_row['Temperature']:.2f}°C")
print(f"Tarih: {max_temp_row['Date']}")
print(f"Şehir: {max_temp_row['City']}")
print("\n" + "="*70 + "\n")
```

```python
# En düşük nem oranının olduğu satır
min_humidity_idx = df['Humidity'].idxmin()
min_humidity_row = df.loc[min_humidity_idx]

print("En düşük nem oranının olduğu satır:")
print(min_humidity_row)
print(f"\nEn düşük nem: {min_humidity_row['Humidity']:.2f}%")
print(f"Tarih: {min_humidity_row['Date']}")
print(f"Şehir: {min_humidity_row['City']}")
```

## 10. Dışa Aktarma (Export)

Şehirlere göre ortalama sıcaklık tablosunu Excel dosyası olarak kaydediyoruz.

```python
# Şehirlere göre ortalama sıcaklığı DataFrame'e çeviriyoruz
city_temp_summary = df.groupby('City')['Temperature'].agg([
    ('Ortalama Sıcaklık', 'mean'),
    ('Maksimum Sıcaklık', 'max'),
    ('Minimum Sıcaklık', 'min'),
    ('Kayıt Sayısı', 'count')
]).round(2)

# DataFrame'i sırala
city_temp_summary = city_temp_summary.sort_values('Ortalama Sıcaklık', ascending=False)

print("Şehirlere göre sıcaklık özeti:")
print(city_temp_summary)
print("\n" + "="*70 + "\n")
```

```python
# Excel dosyası olarak kaydetme
try:
    city_temp_summary.to_excel('sehir_sicakliklari.xlsx')
    print("✓ Dosya başarıyla kaydedildi: sehir_sicakliklari.xlsx")
except Exception as e:
    # Excel yoksa CSV olarak kaydet
    city_temp_summary.to_csv('sehir_sicakliklari.csv')
    print("✓ Excel kütüphanesi bulunamadı. CSV olarak kaydedildi: sehir_sicakliklari.csv")
    print(f"Hata: {e}")
```
